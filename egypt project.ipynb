```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
import os

folder_name='csvs'
csv_list=[]
for file_name in os.listdir(folder_name):
    csv_list.append(pd.read_csv(folder_name+'/'+file_name))
```

```python
list_names=[]
for i in range(len(csv_list)):
    name=csv_list[i]['Indicators'][0]
    list_names.append(name)
list_names
```

```python
import matplotlib.ticker as tick

def transform_axis_fmt(tick_val, pos):
    if  tick_val >= 1000000: 
        val = int(tick_val/1000000)
        return '{:d}M'.format(val)
    elif tick_val >= 1000:
        val = int(tick_val/1000)
        return '{:d}K'.format(val)
    else:
        return int(tick_val)
```

```python
plt.figure(figsize=[20,6])
sb.pointplot(data=total_employment[total_employment['!DownloadDisplay'].notnull()].sort_values(by='Year'),
             x='Year',y='!DownloadDisplay')
plt.title('Total Employment', fontsize=15)
plt.xlabel('Year',fontsize=12)
plt.xticks(rotation=60)
plt.ylabel('',fontsize=12)
plt.gca().yaxis.set_major_formatter(tick.FuncFormatter(transform_axis_fmt));
```

## Exchange Rate

```python
cny_usd=pd.read_csv('/Users/xuanyinghe/Desktop/Github/EgyptProject/Data_Exchange_Rate_CNY_USD.csv')
cny_usd.head(3)
```

```python
egp_usd=pd.read_csv('/Users/xuanyinghe/Desktop/Github/EgyptProject/Data_Exchange_Rate_EGP_USD.csv')
egp_usd.head(3)
```

```python
egp_usd=egp_usd.iloc[2:].reset_index(drop=True)
```

```python
egp_usd.head(3)
```

```python
egp_usd['Date']=pd.to_datetime(egp_usd['Date'])
cny_usd['Date']=pd.to_datetime(cny_usd['Date'])
```

```python
egp_usd=egp_usd[egp_usd['Date'].dt.year>=2010]
cny_usd=cny_usd[cny_usd['Date'].dt.year>=2010]
```

```python
egp_cny=pd.merge(egp_usd,cny_usd,on='Date')
egp_cny.head(3)
```

```python
egp_cny['EGP_CNY']=egp_cny['EGP=X.Adjusted']/egp_cny['CNY=X.Adjusted']
egp_cny.head(3)
```

```python
egp_cny.plot(kind='line',x='Date',y='EGP_CNY');
```
